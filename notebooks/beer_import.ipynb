```python
# Function for flattening  
# json 
def flatten_json(y): 
    out = {} 
  
    def flatten(x, name =''): 
          
        # If the Nested key-value  
        # pair is of dict type 
        if type(x) is dict: 
              
            for a in x: 
                flatten(x[a], name + a + '_') 
                  
        # If the Nested key-value 
        # pair is of list type 
        elif type(x) is list: 
              
            i = 0
              
            for a in x:                 
                flatten(a, name + str(i) + '_') 
                i += 1
        else: 
            out[name[:-1]] = x 
  
    flatten(y) 
    return out 
  
# Driver code 
# print(flatten_json(unflat_json))  
```

```python
# Grab breweries
brewery_data = []
breweries_page_num = 1
breweries_start_url = url+'breweries'+real_key+page+str(breweries_page_num)
breweries_next_page = breweries_start_url

# While loop to grab all pages
while breweries_next_page:
    breweries_resp = requests.get(breweries_next_page)
#     print(resp.status_code)
#   turn request into a json object. Will not be flatten
    breweries_unflat_json =  breweries_resp.json()
    if int(breweries_unflat_json['currentPage']) < int(breweries_unflat_json['numberOfPages'])+1:
        brewery = pd.json_normalize(breweries_unflat_json,'data', max_level=3)
        brewery_data.append(brewery)
        breweries_page_num = breweries_page_num + 1
#       reassign next_page back to itself to update page number
        breweries_next_page = url+'breweries'+real_key+page+str(breweries_page_num)
#         print(next_page)
        
    else:
        print('end')
        break
```

```python
brewery_df = pd.concat(brewery_data)
# brewery_data
```

```python
# Grab locations
locations_data = []
locations_page_num = 1
locations_start_url = url+'locations'+real_key+page+str(locations_page_num)
locations_next_page = locations_start_url

# While loop to grab all pages
while locations_next_page:
    locations_resp = requests.get(locations_next_page)
    print(locations_resp.status_code)
#     print(locations_data)
#   turn request into a json object. Will not be flatten
    locations_unflat_json =  locations_resp.json()
    if int(locations_unflat_json['currentPage']) < int(locations_unflat_json['numberOfPages'])+1:
        location = pd.json_normalize(locations_unflat_json,'data', max_level=3)
        locations_data.append(location)
        locations_page_num = locations_page_num + 1
#       reassign next_page back to itself to update page number
        locations_next_page = url+'locations'+real_key+page+str(locations_page_num)
#         print(next_page)
        
    else:
        print('end')
        break
```

```python
location_df = pd.concat(locations_data)
```
