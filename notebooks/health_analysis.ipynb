```python
# Create state column so all counties are mapped with full state name
health_indicators['state'] = health_indicators.state_abbreviation.map(states) 
```

```python
# merge both df so region and division are added
health_final = pd.merge(health_indicators, region_division, left_on='state', right_on='name')
```

```python
# map the regions so numbers are the name of the region
health_final['region'] = health_final.region.map(region_map) 
```

```python
# map the divisions so numbers are the name of the divsions
health_final['division'] = health_final.division.map(division_map) 
```

```python
health_final = health_final.drop(['name'], axis = 1)
```
